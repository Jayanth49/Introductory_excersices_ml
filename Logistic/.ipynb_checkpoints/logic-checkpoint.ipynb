```python
import numpy as np
import  matplotlib.pyplot as plt
import pandas as pd
```

```python
d = pd.read_csv('ds1_train.csv').values
x_1,x_2 = np.matrix(d[:,0],dtype=float).T,np.matrix(d[:,1],dtype=float).T
y = np.matrix(d[:,2],dtype=float).T
n = x_1.shape[0]
X = np.concatenate([x_1,x_2,np.ones((n,1))],axis=1)
d.shape
x_3 = d[:,0]
x_4 = d[:,1]
y_1 = d[:,2]
plt.scatter(x_3,x_4,c=y_1)
y.shape
```

```python
def sigmond(x):
    t = 1/(1+np.exp(-x))
    return t
```

```python
def predicted_values(X,theta):
    x = X@theta
    return sigmond(x)
```

```python
def loss(predicted,y):
    return predicted-y
```

```python
def gradient(X,los,alpha):
    m = X.shape[0]
    grad = X.T@los
    return (1/m)*alpha*grad
```

```python
def update(grad,theta):
    return theta-grad
```

```python
def gradient_descent(X,y,theta,num_iters,alpha):
    for i in range(num_iters):
        predicted = predicted_values(X,theta)
        los = loss(predicted,y)
        grad = gradient(X,los,alpha)
        theta = update(grad,theta)
    return theta
```

```python
theta = np.matrix([0,0,0]).T
theta = gradient_descent(X,y,theta,80500,0.001)
#error = predicted_values(X,theta)-y
theta
```
